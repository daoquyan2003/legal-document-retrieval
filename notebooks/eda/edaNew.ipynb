```python
import json
import matplotlib.pyplot as plt

# Load data from JSON files
with open('train.json', encoding="utf8") as f:
    train_data = json.load(f)
with open('additional_data/zalo/zalo_question.json', encoding="utf8") as f:
    train_zalo_data = json.load(f)
with open('additional_data/ALQAC_2022_training_data/question.json', encoding="utf8") as f:
    train_alqac2022_data = json.load(f)
# with open('E:/Nm 3 - K 1/X l ngn ng t nhin INT3408E 20/nlp/legal-document-retrieval/data/ALQAC_2023_training_data_v1.1/additional_data/ALQAC_2021/question.json', encoding="utf8") as f:
#     train_alqac2021_data = json.load(f)   
with open('additional_data/zalo/zalo_corpus.json', encoding="utf8") as f:
    law_zalo_data = json.load(f)
with open('law.json', encoding="utf8") as f:
    law_original_data = json.load(f)
with open('additional_data/ALQAC_2022_training_data/law.json', encoding="utf8") as f:
    law_alqac2022_data = json.load(f)
# with open('E:/Nm 3 - K 1/X l ngn ng t nhin INT3408E 20/nlp/legal-document-retrieval/data/ALQAC_2023_training_data_v1.1/additional_data/ALQAC_2021/law.json', encoding="utf8") as f:
#     law_alqac2021_data = json.load(f)


# Extract word lengths
question_lengths_original = [len(item['text'].split()) for item in train_data]
question_lengths_zalo = [len(item['text'].split()) for item in train_zalo_data]
question_lengths_alqac2022 = [len(item['text'].split()) for item in train_alqac2022_data]
# question_lengths_alqac2021 = [len(item['text'].split()) for item in train_alqac2021_data]

article_lengths_original = [len(article['text'].split()) for law in law_original_data for article in law['articles']]
article_lengths_zalo = [len(article['text'].split()) for law in law_zalo_data for article in law['articles']]
article_lengths_alqac2022 = [len(article['text'].split()) for law in law_alqac2022_data for article in law['articles']]
# article_lengths_alqac2021 = [len(article['text'].split()) for law in law_alqac2021_data for article in law['articles']]

# Create histograms
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
plt.hist(question_lengths_original, bins=200, range=(0, 200), color='blue', alpha=0.7)
plt.title('Question Length - ALQAC 2023')

plt.subplot(2, 3, 2)
plt.hist(question_lengths_alqac2022, bins=200, range=(0, 200), color='orange', alpha=0.7)
plt.title('Question Length - ALQAC 2022')

# plt.subplot(2, 4, 3)
# plt.hist(question_lengths_alqac2021, bins=200, range=(0, 200), color='red', alpha=0.7)
# plt.title('Question Length - ALQAC 2021')

plt.subplot(2, 3, 3)
plt.hist(question_lengths_zalo, bins=200, range=(0, 200), color='green', alpha=0.7)
plt.title('Question Length - Zalo')

plt.subplot(2, 3, 4)
plt.hist(article_lengths_original, bins=500, range=(0, 2000), color='blue', alpha=0.7)
plt.title('Article Length - AlQAC 2023')


plt.subplot(2, 3, 5)
plt.hist(article_lengths_alqac2022, bins=500, range=(0, 2000), color='orange', alpha=0.7)
plt.title('Article Length - ALQAC 2022')

# plt.subplot(2, 4, 7)
# plt.hist(article_lengths_alqac2021, bins=500, range=(0, 2000), color='red', alpha=0.7)
# plt.title('Article Length - AlQAC 2021')


plt.subplot(2, 3, 6)
plt.hist(article_lengths_zalo, bins=500, range=(0, 2000), color='green', alpha=0.7)
plt.title('Article Length - Zalo')

plt.tight_layout()
plt.show()
```
